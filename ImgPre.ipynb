```python
####################################################
          # Training Date Set 데이터 증식 #
####################################################
import glob 
from IPython.display import Image, display, clear_output
import numpy as np
import pandas as pd
from keras.preprocessing.image import ImageDataGenerator, array_to_img, img_to_array, load_img
from keras.applications.imagenet_utils import preprocess_input, decode_predictions
import matplotlib.pyplot as plt

from keras.preprocessing.image import ImageDataGenerator

datagen = ImageDataGenerator(
        rotation_range=20,
        width_shift_range=0.2,
        height_shift_range=0.2,
        shear_range=0.1,
        zoom_range=0.2,
        horizontal_flip=True,
        fill_mode='nearest')

# 파손 이미지
images1 = glob.glob('how2/crash/*.jpeg')
images2 = glob.glob('how2/crash/*.jpg')
images3 = glob.glob('how2/crash/*.png')
images = images1 + images2 + images3

for j in range(len(images)):
        img = load_img(images[j])  # PIL 이미지
        x = img_to_array(img)  # (3, 150, 150) 크기의 NumPy 배열
        x = x.reshape((1,) + x.shape)  # (1, 3, 150, 150) 크기의 NumPy 배열

        # 아래 .flow() 함수는 임의 변환된 이미지를 배치 단위로 생성해서
        # 지정된 `preview/` 폴더에 저장합니다.
        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                save_to_dir='how2_train/crash', save_prefix='agd', save_format='jpeg'):
                i += 1
                if i > 3:
                        break  # 이미지 3장을 생성하고 마칩니다
                        
# 파손 이미지
images4 = glob.glob('how2/dent/*.jpeg')
images5 = glob.glob('how2/dent/*.jpg')
images6 = glob.glob('how2/dent/*.png')
images = images4 + images5+ images6 

for j in range(len(images)):
        img = load_img(images[j])  # PIL 이미지
        x = img_to_array(img)  # (3, 150, 150) 크기의 NumPy 배열
        x = x.reshape((1,) + x.shape)  # (1, 3, 150, 150) 크기의 NumPy 배열

        # 아래 .flow() 함수는 임의 변환된 이미지를 배치 단위로 생성해서
        # 지정된 `preview/` 폴더에 저장합니다.
        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                save_to_dir='how2_train/dent', save_prefix='agd', save_format='jpeg'):
                i += 1
                if i > 3:
                        break  # 이미지 3장을 생성하고 마칩니다
                        
# 파손 이미지
images7 = glob.glob('how2/glb/*.jpeg')
images8 = glob.glob('how2/glb/*.jpg')
images9 = glob.glob('how2/glb/*.png')
images = images7 + images8+ images9 

for j in range(len(images)):
        img = load_img(images[j])  # PIL 이미지
        x = img_to_array(img)  # (3, 150, 150) 크기의 NumPy 배열
        x = x.reshape((1,) + x.shape)  # (1, 3, 150, 150) 크기의 NumPy 배열

        # 아래 .flow() 함수는 임의 변환된 이미지를 배치 단위로 생성해서
        # 지정된 `preview/` 폴더에 저장합니다.
        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                save_to_dir='how2_train/glb', save_prefix='agd', save_format='jpeg'):
                i += 1
                if i > 3:
                        break  # 이미지 3장을 생성하고 마칩니다
                        
# 파손 이미지
images10 = glob.glob('how2/scratch/*.jpeg')
images11 = glob.glob('how2/scratch/*.jpg')
images12 = glob.glob('how2/scratch/*.png')
images = images10 + images11+ images12

for j in range(len(images)):
        img = load_img(images[j])  # PIL 이미지
        x = img_to_array(img)  # (3, 150, 150) 크기의 NumPy 배열
        x = x.reshape((1,) + x.shape)  # (1, 3, 150, 150) 크기의 NumPy 배열

        # 아래 .flow() 함수는 임의 변환된 이미지를 배치 단위로 생성해서
        # 지정된 `preview/` 폴더에 저장합니다.
        i = 0
        for batch in datagen.flow(x, batch_size=1,
                                save_to_dir='how2_train/scratch', save_prefix='agd', save_format='jpeg'):
                i += 1
                if i > 3:
                        break  # 이미지 3장을 생성하고 마칩니다
```
